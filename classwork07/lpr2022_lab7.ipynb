```python
import pandas as pd

titanic = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv")

titanic
```

```python
titanic.groupby('class').agg({"who":"count"})
```

```python
titanic.groupby('age').agg({"who":"count"})
```

```python
t_not_null = titanic.dropna(subset=['age'], how='all')
def sep(pclass, sex):
    age_grop = t_not_null[(t_not_null['pclass']==pclass)&(t_not_null['sex']==sex)].sort_values(by="age")
    age_grop1 = age_grop[:len(age_grop)//2]
    age_grop2 = age_grop[len(age_grop)//2:]
    return age_grop1, age_grop2
num = [(1,'male'), (2,'male'), (3,'male'), (1,'female'), (2,'female'), (3,'female')]
groups = []
for pclass, sex in num:
    g1, g2 = sep(pclass, sex)
    groups.append(g1)
    groups.append(g2)
print(*groups)
```

```python
survived = []
for g in groups:
    a = len(g[g.survived == 1].index.values)/len(g.index.values)
    survived.append(a)
print(*survived)
```

```python
age_alive_male = t_not_null[t_not_null.sex == 'male'].groupby('survived').agg({'age':'mean'}).loc[1]
age_alive_female = t_not_null[t_not_null.sex == 'female'].groupby('survived').agg({'age':'mean'}).loc[1]
print(age_alive_male)
print(age_alive_female)
```

```python
t_not_null[t_not_null['survived']==1]['age'].std()
```

```python
import matplotlib.pyplot as plt

t_age = t_not_null.sort_values(by="age")
x = t_age.groupby('age').mean().index.values
y = t_age.groupby('age').mean().survived.values
plt.plot(x, y)
```
